# Matplotlib - 기본 plot

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.plot([0,1,4,9])     # X, Y (X가 생략되면 0,1,2,3, ... 을 의미)
plt.title('라인 플롯')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')        # color, marker, linestyle / 'ro-' == red / o(circle) / -(line)   
plt.title('Red-Circle-직선')  
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--')       # color, marker, linestyle
plt.plot([8,4,2,1], 'g^:')
plt.title('Magenta-Square-Dashed, Green-삼각형-점선')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--', linewidth=3, markersize=10)       
plt.plot([8,4,2,1], 'g^:', lw=5, ms=20)
plt.title('Magenta-Square-Dashed, Green-삼각형-점선')
plt.grid()
plt.show()
```

- 그림 범위

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')
plt.title('X축, Y축 범위 지정')
plt.xlim(-1,4); plt.ylim(-1,10)     # ;(세미콜론) ==> 여러개 코드 한 줄에!
plt.grid()
plt.show()  
```

- 곡선 그림

```python
X = np.arange(0, 2*np.pi, 0.01)
Ysin, Ycos = np.sin(X), np.cos(X) 
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Sine & Cosine 그래프')
plt.grid(); plt.show()
```

- Tick 설정

```python
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Tick 설정')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$', r'$-\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTex     // 데이터사이언스스쿨 참조. . .
plt.yticks([-1, 0, 1])
plt.grid(); plt.show()
```

- 범례(Legend)

```python
plt.plot(X, Ysin, label='sine'), plt.plot(X, Ycos, label='cosine')
plt.title('범례(Legend) 표시')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$', r'$-\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])
plt.yticks([-1, 0, 1])    
plt.legend(); plt.grid(); plt.show()
```

- X축, Y축 Label

```python
plt.plot(X, Ysin, label='sine'), plt.plot(X, Ycos, label='cosine')
plt.title('X축, Y축 Label 표시', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$', r'$-\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])   
plt.yticks([-1, 0, 1])
plt.xlabel('시간'), plt.ylabel('진폭') 
plt.legend(); plt.grid(); plt.show()
```

- 그림 크기 지정

```python
plt.figure(figsize=(10,6))
plt.plot(X, Ysin, label='sine'), plt.plot(X, Ycos, label='cosine')
plt.title('그림 크기 지정(10,6)', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$', r'$-\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])   
plt.yticks([-1, 0, 1])
plt.xlabel('시간'), plt.ylabel('진폭') 
plt.legend(); plt.grid(); plt.show()
```

- 그림 저장

```python
plt.figure(figsize=(10,6))
plt.plot(X, Ysin, label='sine'), plt.plot(X, Ycos, label='cosine')
plt.title('그림 저장', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$', r'$-\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])   
plt.yticks([-1, 0, 1])
plt.xlabel('시간'), plt.ylabel('진폭') 
plt.legend(); plt.grid()
plt.savefig('삼각함수.png')
```
